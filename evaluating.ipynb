```python
# Задание 2. Отберите худших врачей по каждой из патологий. 
# Комментарий: Необходимо описать принцип отбора
```

```python
# Для определения несогласованности врача с другими
# рассчитаем коэффициент Жаккарда
# Чем ниже этот коэффициент, тем больше несогласованность врача с другими
```

```python
hydro_data = pd.read_csv('Гидроторакс.csv')
ochag_data = pd.read_csv('Очаг.csv')
pneumo_data = pd.read_csv('Снижение пневматизации.csv')
```

```python
# Вычисляем коэффициент Жаккарда
def jaccard_ind(doctor_col, majority_col):
    agree = len(doctor_col[doctor_col == majority_col])
    return agree / len(doctor_col)
```

```python
def find_worst_doctors(data, disease, num_worst=3):
    majority_col = data.apply(lambda row: row.value_counts().index[0], axis=1)
    worst_doctors = []
    for doctor in data.columns[1:]:
        jaccard = jaccard_ind(data[doctor], majority_col)
        worst_doctors.append((doctor, jaccard))
    worst_doctors.sort(key=lambda x:x[1])
    print(f'\nНаиболее несогласные врачи по патологии {disease}: ')
    for i in range(num_worst):
        print(f'{worst_doctors[i][0]}: {worst_doctors[i][1]}')
find_worst_doctors(hydro_data, "Гидраторакс")
find_worst_doctors(ochag_data, "Очаг")
find_worst_doctors(pneumo_data, "Снижение пневматизации")
```

```python
# Задание 3. Сформируйте итоговую разметку (для каждого снимка только одно значение 0 или 1).
# А также уверенность в разметке каждого снимка.
# Комментарий: критерий уверенности в разметке предлагается выбрать самим
```

```python
def get_final_labels(filename):
    df = pd.read_csv(filename)

    final_df = pd.DataFrame(columns=['ID Файла', 'Итоговая разметка', 'Уверенность'])

    for file_id in df['ID Файла'].unique():

        file_labels = df[df['ID Файла'] == file_id].drop('ID Файла', axis=1).values.flatten()

        positive_votes = np.sum(file_labels)
        negative_votes = len(file_labels) - positive_votes
        # Определяем итоговую разметку
        final_label = 1 if positive_votes > negative_votes else 0
        # Определяем уверенность
        confidence = positive_votes / len(file_labels) if final_label == 1 else negative_votes / len(file_labels)
        
        
        new_row = pd.DataFrame({'ID Файла': [file_id], 'Итоговая разметка': [final_label], 'Уверенность': [confidence]})
        final_df = pd.concat([final_df, new_row], ignore_index=True)
    return final_df

hydra_df = get_final_labels('Гидроторакс.csv')
ochag_df = get_final_labels('Очаг.csv')
pneumo_df = get_final_labels('Снижение пневматизации.csv')
```

```python
# Задание 4. Оцените итоговое качество разметки по каждой из патологий.
# Комментарий: необходимо получить численную оценку,
# а также можно привести комментарии с доводами о качестве разметки
```

```python
from collections import Counter

def calculate_kappa(df):

  # Снимки, где врачи дали одиковый результат
  n_total = len(df)
  n_agree = 0
  for i in range(n_total):
    if len(set(df.iloc[i].values)) == 1:
      n_agree += 1

  # Согласованность врачей
  po = n_agree / n_total

  # Ожидаемая согласованность
  pe = 0
  for col in df.columns:
    n_1 = df[col].sum()
    n_0 = len(df) - n_1
    pe += (n_1 / n_total) * (n_1 / n_total) + (n_0 / n_total) * (n_0 / n_total)
  pe /= len(df.columns)
 

  # Коэффициент Kappa
  kappa = (po - pe) / (1 - pe)
  return kappa


df = pd.read_csv('Гидроторакс.csv', index_col='ID Файла')
kappa = calculate_kappa(df)
print(f'Kappa для Гидроторакс: {kappa:.3f}')

df = pd.read_csv('Очаг.csv', index_col='ID Файла')
kappa = calculate_kappa(df)
print(f'Kappa для Очаг: {kappa:.3f}')

df = pd.read_csv('Снижение пневматизации.csv', index_col='ID Файла')
kappa = calculate_kappa(df)
print(f'Kappa для Снижение пневматизации: {kappa:.3f}')
```

```python
# Полученные значения коэффициента Kappa показывают,
# Что наблюдаемая согласованность ниже, чем ожидаемая при случайном совпадении,
# Это говорит о том, что врачи скорее расходятся во мнении, чем сходятся,
# Даже если учитывать случайные совпадения
```

```python

hydra = pd.read_csv("Гидроторакс.csv")
ochag = pd.read_csv("Очаг.csv")
pneumo = pd.read_csv("Снижение пневматизации.csv")

# Объединение разметки
def get_label(row):
    votes = row.values[1:]
    if votes.sum() >= 5:
        return 1
    else:
        return 0

hydra["label"] = hydra.apply(get_label, axis=1)
ochag["label"] = ochag.apply(get_label, axis=1)
pneumo["label"] = pneumo.apply(get_label, axis=1)

# Сравнение с итоговой разметкой
expert1 = hydra_df['Итоговая разметка']
expert2 = ochag_df['Итоговая разметка']
expert3 = pneumo_df['Итоговая разметка']

# Расчет метрик
def calculate_metrics(df, expert):
    y_true = expert
    y_pred = df["label"]
    tp = sum((y_true == 1) & (y_pred == 1))
    fp = sum((y_true == 0) & (y_pred == 1))
    fn = sum((y_true == 1) & (y_pred == 0))

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1

hydra_metrics = calculate_metrics(hydra, expert1)
ochag_metrics = calculate_metrics(ochag, expert2)
pneumo_metrics = calculate_metrics(pneumo, expert3)

# Вывод результатов
print("Гидроторакс: (Precision, recall, f1)\n", hydra_metrics)
```
